# Summary of Data
The  main focus of our project is converting hand-written english text into an audio sample of the given text. To train our models, all we need to do is get a variety of hand-writing from different people so that we can have a diverse test group based on different styles of writing. We have 3 different sets of handwriting to start, as we need to train the model to discern between the different english characters with different handwriting styles. Once the model can discern english characters, we just need to append the detected characters together to form a sentence. A sentence also requires punctions, which is why our model will also be trained on english punctuation.
